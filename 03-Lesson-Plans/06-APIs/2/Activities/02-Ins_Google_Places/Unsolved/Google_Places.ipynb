```python
# Dependencies
import requests
import json

# Google developer API key
from config import gkey
```

```python
# geocoordinates
target_coordinates = "43.6187102, -116.2146068"
target_search = "Chinese"
target_radius = 8000
target_type = "restaurant"
```

```python
target_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
##build Params dictionaries

params ={
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": gkey    
}
```

```python
## API request

response = requests.get(target_url,params=params)
```

```python
# how we check we got a response
response
```

```python
#how we see what the response is
response = response.json()
```
